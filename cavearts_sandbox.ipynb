```python
import pandas as pd
```

```python
input_folder = "./input/"

data = pd.read_csv(input_folder+'africa-economic-banking-and-systemic-crisis-data.zip', compression='zip')
```

```python
if not data.isna().any().any():
    print("There are no NaN values in the dataset!")
```

```python
import plotly.graph_objects as go

```

```python
fig = go.Figure(data=go.Choropleth(
    locations=data['cc3'],
    z=data['exch_usd'].astype(float),
    locationmode = 'country names',
    colorscale='Reds',
    autocolorscale=False,
    text=data['banking_crisis'], # hover text
    marker_line_color='white', # line markers between states
    colorbar_title="Debt in Millions USD"
))

fig.update_layout(
    title_text='Crisis Town yo!',
    geo = dict(
        scope='africa',
        projection = go.layout.geo.Projection(type = 'equirectangular'),
        showlakes=True, # lakes
        lakecolor='rgb(255, 255, 255)'),
)

fig.show()
```
